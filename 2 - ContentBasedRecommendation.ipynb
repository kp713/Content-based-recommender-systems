### Content Based Recommendation System

```python
import pandas as pd

df = pd.read_csv('2 - movies_metadata - Content based Recommendation.csv')
```

```python
df
```

```python
df['tagline'].fillna('')
df['description'] = df['overview'] + df['tagline'] #new col called description to use in Bow
# df['description'] = df['description'].fillna('')
```

```python
df.shape
```

```python
df.dropna(subset=['title'], inplace=True)
df.dropna(subset=['description'], inplace=True)
df['title'].drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.reset_index()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=2, stop_words='english') #analyzer=alphabet every alphabet becomes a unique col
tfidf_matrix = tf.fit_transform(df['description'])
print(tfidf_matrix)
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

print(ENGLISH_STOP_WORDS)
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_similarities.shape
```

```python
cosine_similarities[0].shape
```

```python
df = df.reset_index()
titles = df['title']
titles.shape
```

```python
indices = pd.Series(df.index, index=df['title'])
```

```python
def recommend(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_similarities[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```
